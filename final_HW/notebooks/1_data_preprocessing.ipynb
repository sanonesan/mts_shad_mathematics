## Обработка данных

Заметим, что в датасете первое слово в колонке `name` обозначает производителя, добавим это как фитчу.

```python
def add_brand(df):
    df["brand"] = df["name"].apply(lambda x: x.split(" ", 1)[0])

add_brand(df_train)
add_brand(df_test)
```

```python
df_train.head()
```

Преобразуем колонки `mileage`, `max_power`, `engine` из string во float, чтобы остались только значения kmpl, bhp, CC соответственно

```python
def clean_mileage_maxpower_engine(df):

    def clean_str_to_float(x):
        res = str(x).split(" ", 1)[0]
        if res == "":
            return np.nan
        else:
            return np.float64(res)

    df["mileage"] = df["mileage"].apply(clean_str_to_float)
    df["max_power"] = df["max_power"].apply(clean_str_to_float)
    df["engine"] = df["engine"].apply(clean_str_to_float)

    df = df.dropna()

clean_mileage_maxpower_engine(df_train)
clean_mileage_maxpower_engine(df_test)

df_train = df_train.dropna().drop_duplicates()
df_test = df_test.dropna().drop_duplicates()
```

```python
df_train.head()
```

Разбираемся с колонкой `torque` (крутящий момент)

```python
def clean_torque(df):

    df["max_torque"] = df["torque"].str.extract(
        "([\d.]+)Nm",
        flags=re.IGNORECASE,
    )

    unprocessed_df = df[df["max_torque"].isna() & df["torque"].notna()]

    for index, item in unprocessed_df.iterrows():
        # обработаем строки вида 11.5@ 4,500(kgm@ rpm) и 22.4 kgm at 1750-2750rpm	
        res = re.findall("^[\d.]+", item["torque"], flags=re.IGNORECASE)
        if res:
            res = float(res[0])
            if "kgm" in item["torque"].lower():
                # переводим из kgm в Nm
                res *= 9.80665
            df.loc[index, "max_torque"] = res

    df["max_torque"] = df["max_torque"].astype(float)

    def torque_rpm_agg(x):
        if len(x) == 0:
            return np.nan
        for x in x[-1]:
            if x != "":
                return x.replace(",", "")
        return np.nan

    tmp = df["torque"].str.findall(
        "(\d,\d{3,})|(\d{4,})",
    ).apply(torque_rpm_agg)

    # for index, item in tmp[tmp[1].isna()].iterrows():
    #     tmp.loc[index, 1] = tmp.loc[index, 0]

    df["max_torque_rpm"] = tmp.astype(np.float64)

clean_torque(df_train)
clean_torque(df_test)

df_train = df_train.dropna().drop_duplicates()
df_test = df_test.dropna().drop_duplicates()

```

Удалим колонку `torque`

```python
df_train = df_train.drop(columns=["torque"])
df_train
```

```python
df_test = df_test.drop(columns=["torque"])
df_test
```

Переведем некоторые колонки из float в int

```python
def float_to_int(df):
    df["engine"] = df["engine"].astype(np.int32)
    df["seats"] = df["seats"].astype(np.int16)
    df["max_torque"] = df["max_torque"].round().astype(np.int16)
    df["max_torque_rpm"] = df["max_torque_rpm"].astype(np.int32)

float_to_int(df_train)
float_to_int(df_test)
```
